# Merge, join, and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\nhi\OneDrive\Desktop\Data Analyst\PythonPandas\LOTR.csv")
```

```python
df1
```

```python
df2 = pd.read_csv(r"C:\Users\nhi\OneDrive\Desktop\Data Analyst\PythonPandas\LOTR 2.csv")
```

```python
df2
```

```python
# merge function-- inner join
df1.merge(df2, how = 'inner')
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID') #inner join on FellowshipID column
```

```python
# merge function-- outer join-- return everything from both dataframes

df1.merge(df2, how = 'outer')
```

```python
# merge function-- left join-- return everything from left dataframe

df1.merge(df2, how = 'left')
```

```python
# merge function-- right join-- return everything from right dataframe

df1.merge(df2, how = 'right')
```

```python
# merge function-- cross join-- return each value from left dataframe to every value from the right dataframe

df1.merge(df2, how = 'cross')
```

```python
# join function-- used when work with indexing

df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix='_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix='_Right', how = 'outer')
```

```python
df4
```

```python
# concatenate fucntion-- concat left df on top of right df
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join = 'inner')
```

```python
pd.concat([df1, df2], join = 'outer')
```

```python
pd.concat([df1, df2], join = 'outer', axis=1)
```

# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\nhi\OneDrive\Desktop\Data Analyst\PythonPandas\Customer Call List.xlsx")
```

```python
df
```

```python
# drop_duplicates function

df = df.drop_duplicates()
```

```python
df
```

```python
# drop function
# shift + tab to see the parameters
df = df.drop(columns ="Not_Useful_Column")
df
```

```python
# strip function to clean up the unesscessary character before or after a string

# only apply strip function to Last_Name column in dataframe
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
#df

df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df
```

```python
# replace function
# To get phone number into format xxx-xxx-xxxx
# apply replace function to Phone_Number column

#df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)) #convert  each row of Phone_Number column into a string
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('N/a--', '')

df["Phone_Number"] = df["Phone_Number"].str.replace('|', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('-', '')
```

```python
# apply for format xxx-xxx-xxxx
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
# split function
# split Address column where comma is 
df [["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True )
df
```

```python
# replace Paying Customer column in format Y or N
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')

df
```

```python
# replace Do_Not_Contact column in format Y or N
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')

df
```

```python
# replace N/a of entire dataframe to ''
# df = df.replace('N/a','')
```

```python
# fillna function- fill every non-value to empty string
df = df.fillna('')
```

```python
df
```

```python
# drop the row that has Y in Do_not_contact
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
        
df
```

```python
# drop the row that has no phone number in Phone_Number
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x, inplace=True)
        
df

# Another way to drop null value
# df = df.dropna(subset="Phone_Number", inplace=True)
```

```python
df = df.reset_index(drop=True) # drop original index
```
